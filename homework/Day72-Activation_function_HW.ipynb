# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

def ReLU(x):
    return abs(x)*(x>0)
def dReLU(x):
    return (1*(x>0))
```

```python
x=plt.linspace(-10,10,100)
plt.plot(x,ReLU(x),'r')
plt.text(-2.5, 8, 'ReLU:f(x)=max(0,x)', fontsize=15,color='red')
plt.plot(x,dReLU(x),'b')
plt.text(2.5, 2, 'dReLU: x < 0 ,x = 0',fontsize=12,color='blue')
plt.text(5.5, 1.5, 'x >= 0 ,x = 1',fontsize=12,color='blue')
```
